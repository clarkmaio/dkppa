```python
import arraylake
import xarray as xr
import yaml
import os
from dotenv import load_dotenv
from datetime import datetime
import polars as pl

load_dotenv()
```

```python
token = os.getenv('EARTHMOVER_TOKEN')
client = arraylake.Client(token=token)
print(token)
```

```python
repo_id = "myexperiment/ecmwf-era5"
repo = client.get_repo(repo_id)
session = repo.readonly_session(branch="main")
ds = xr.open_zarr(session.store, zarr_format=3, group='temporal')
lat_min, lat_max, lon_min, lon_max = float(os.environ['MIN_LAT']), float(os.environ['MAX_LAT']), float(os.environ['MIN_LON']), float(os.environ['MAX_LON'])

sliced_ds = ds.sel(
            latitude=slice(lat_max, lat_min),
            longitude=slice(lon_min, lon_max),
            time=slice(datetime(2023, 1, 1), datetime(2023, 1, 1))
        )[['u100', 'v100']]
```

```python
sliced_df = sliced_ds.to_dataframe()
```

```python
sliced_df_pl = pl.from_pandas(sliced_df.reset_index())
sliced_df_pl
```
